```python
import matplotlib.pyplot as plt
import pandas as pd
import glob

# Pattern to match the CSV files
file_pattern = '/home/scratch/jcorner1/syn_sev/dataframes/EOC8p5_UVV*.csv'
files = glob.glob(file_pattern)

# Read and concatenate all the matching files
data_frames = [pd.read_csv(file) for file in files]
data = pd.concat(data_frames)

# Convert 'Time' column to datetime, extract the year and month
data['Year'] = pd.to_datetime(data['Time']).dt.year
data['Month'] = pd.to_datetime(data['Time']).dt.month

# Function to map month to season
def month_to_season(month):
    if month in [12, 1, 2]:
        return 'DJF'
    elif month in [3, 4, 5]:
        return 'MAM'
    elif month in [6, 7, 8]:
        return 'JJA'
    elif month in [9, 10, 11]:
        return 'SON'

# Map the months to seasons
data['Season'] = data['Month'].apply(month_to_season)

# Count occurrences of each year per season and store as a list
season_values = [data[data['Season'] == season]['Year'].value_counts() for season in ['DJF', 'MAM', 'JJA', 'SON']]

# Plot the frequencies using a boxplot
fig = plt.figure(figsize =(10, 7))
bp = plt.boxplot(season_values, labels=['DJF', 'MAM', 'JJA', 'SON'])

# Calculate and plot high and low extremes with transparency
for i, season_data in enumerate(season_values, start=1):
    high_extreme = max(season_data)
    low_extreme = min(season_data)
    plt.plot([i], [high_extreme], marker='o', markersize=10, alpha=0.3, color='black')
    plt.plot([i], [low_extreme], marker='o', markersize=10, alpha=0.3, color='black')

plt.title('Boxplot of Counts by Season')
plt.ylabel('Counts')
plt.xlabel('Season')

# Show plot
plt.savefig('boxplot_counts_season.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
import numpy as np
import matplotlib.patches as mpatches
```

```python
paths = {
    "HIST": '/home/scratch/jcorner1/syn_sev/dataframes/HIST_UVV_[21]*_*.csv',
    "EOC_8p5": '/home/scratch/jcorner1/syn_sev/dataframes/EOC8p5_UVV*.csv',
    "MC_8p5": '/home/scratch/jcorner1/syn_sev/dataframes/MC8p5_UVV*.csv',
    "EOC_4p5": '/home/scratch/jcorner1/syn_sev/dataframes/EOC4p5_UVV*.csv',
    "MC_4p5": '/home/scratch/jcorner1/syn_sev/dataframes/MC4p5_UVV*.csv'
}
```

```python
colors = {
    "HIST": 'lime',
    "EOC_8p5": 'lightskyblue',
    "MC_8p5": 'mediumblue',
    "EOC_4p5": 'coral',
    "MC_4p5": 'red'
}

legend_handles = [mpatches.Patch(color=color, label=label) for label, color in colors.items()]
```

```python
def month_to_season(month):
    if month in [12, 1, 2]:
        return 'DJF'
    elif month in [3, 4, 5]:
        return 'MAM'
    elif month in [6, 7, 8]:
        return 'JJA'
    elif month in [9, 10, 11]:
        return 'SON'
```

```python
seasons = ['DJF', 'MAM', 'JJA', 'SON']
data_for_plotting = {season: [] for season in seasons}
```

```python
for name, path in paths.items():
    # Read and concatenate all matching files for the current path
    data_frames = [pd.read_csv(file) for file in glob.glob(path)]
    data = pd.concat(data_frames)
    data = data[(data['UVV'] >= 25) & (data['DBZ'] >= 40)]


    # Convert 'Time' column to datetime, extract year and month
    data['Year'] = pd.to_datetime(data['Time']).dt.year
    data['Month'] = pd.to_datetime(data['Time']).dt.month
    data['Season'] = data['Month'].apply(month_to_season)

    # Append data for each season
    for season in seasons:
        season_data = data[data['Season'] == season]['Year'].value_counts()
        data_for_plotting[season].append((name, season_data))
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

# Plot data grouped by season
width = 0.15  # width of the box plots
for idx, season in enumerate(seasons):
    for j, (name, data) in enumerate(data_for_plotting[season]):
        if not data.empty:
            # Position each dataset within the group for the current season
            position = idx - (len(data_for_plotting[season]) - 1) * width / 2 + j * width
            box = ax.boxplot(data, positions=[position], widths=width, patch_artist=True)
            
            for patch in box['boxes']:
                patch.set_facecolor(colors[name])  # Set the color for each box plot based on the dataset
                
            for median in box['medians']:
                median.set_color('black')
                median.set_linewidth(2)
                
            q1 = np.percentile(data, 25)
            q3 = np.percentile(data, 75)
            iqr = q3 - q1
            lower_whisker = q1 - 1.5 * iqr
            upper_whisker = q3 + 1.5 * iqr

            # Define low and high outliers as min and max values within the data outside whiskers
            low_outlier = data[data < lower_whisker].min()
            high_outlier = data[data > upper_whisker].max()
            
            if not np.isnan(low_outlier):
                ax.scatter(position, low_outlier, color='blue', zorder=3)  # Plot low outlier in blue
            if not np.isnan(high_outlier):
                ax.scatter(position, high_outlier, color='red', zorder=3)  # Plot high outlier in red


ax.set_title('Frequency Changes in UVV ')
ax.set_ylabel('Frequency of Occurrence')
ax.set_xlabel('Season')
ax.set_xticks(np.arange(len(seasons)))
ax.set_xticklabels(seasons)
ax.legend(handles=legend_handles, title="Datasets")

plt.tight_layout()
plt.savefig('Botplot_svrfrq.png',dpi=300)
plt.show()
```
